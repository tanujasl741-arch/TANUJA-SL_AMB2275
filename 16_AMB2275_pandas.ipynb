```python
#filter data (only wheat)
Wheat_data=df[df["Crop"]=="Wheat"]
print(Wheat_data)
#
```

```python
#group by year and sum production
Yearly_production=df.groupby("Year")["Production_Tons"].sum()
print("\nYearly Production_Tons:")
print(Yearly_production)
```

```python
#scatter label-rainfall vs production
import matplotlib.pyplot as plt
plt.scatter(df["Rainfall_mm"],df["Production_Tons"])
plt.xlabel("Rainfall_mm")
plt.ylabel("Production_Tons")
plt.title("Rainfall vs Production")
plt.show()
```

```python
#create barplot
plt.figure(figsize=(5,6))
plt.bar(df["Crop"],df["Production_Tons"],color=['g','y'])
plt.xlabel("Crop")
plt.ylabel("Production_Tons")
plt.title("Crop vs Production")
plt.xticks()
plt.show()
```

```python
#area by yearbar chart
plt.figure(figsize=(4,5))
yr_grp=df.groupby("Year")["Area_Hectares"].sum()
plt.bar(yr_grp.index,yr_grp.values)

plt.xlabel("Area")
plt.ylabel("Year")
plt.title("area vs Year")
plt.xticks()
plt.show()
```

```python
#filter data of each crp
wheat=df[df["Crop"]=="Wheat"]
rice=df[df["Crop"]=="Rice"]
plt.figure()
#plot separately
plt.plot(wheat["Year"],wheat["Production_Tons"],label="wheat")
plt.plot(rice["Year"],rice["Production_Tons"],label="rice")
plt.xlabel("Year")
plt.ylabel("Production_Tons")
plt.title("Crop vs Production")
plt.legend()
plt.show()
```

```python

#custom styled bar plot
plt.figure(figsize=(5,6))
bars=plt.bar(Yearly_production.index,Yearly_production.values,color='b',edgecolor='r')
#add value labels on top of bars
for bar in bars:
  height=bar.get_height()
  plt.text(bar.get_x()+bar.get_width()/2,
           height,
           str(height),
           ha='center',
           va='bottom')
plt.show()
```
