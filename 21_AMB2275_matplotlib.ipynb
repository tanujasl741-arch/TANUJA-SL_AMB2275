```python
#bar plot
import numpy as np
years=[2010,2011,2012,2013,2014,2015,2016,2017]
open_sensex=[1425,1452,1254,1321,1524,1244,1422,1421]
close_sensex=[1525,1512,1354,1421,1624,1444,1521,1231]
ind=np.arange(len(years))
print(ind)
width=0.3
plt.bar(ind,open_sensex,width,label="open")
plt.bar(ind+width,close_sensex,width,label="close",color='g')
plt.title("sensex graph per year")
plt.ylabel("sensex")
plt.xlabel("years")
plt.xticks(ind,years,rotation=70)
plt.legend()
plt.show()
```

```python
#bar plot
import numpy as np
years=[2010,2011,2012,2013,2014,2015,2016,2017]
open_sensex=[1425,1452,1254,1321,1524,1244,1422,1421]
close_sensex=[1525,1512,1354,1421,1624,1444,1521,1231]
ind=np.arange(len(years))
print(ind)
width=0.3
plt.barh(ind,open_sensex,width,label="open")
plt.barh(ind+width,close_sensex,width,label="close",color='g')
plt.title("sensex graph per year")
plt.ylabel("sensex")
plt.xlabel("years")
plt.xticks(ind,years,rotation=70)
plt.legend()
plt.show()
```
