```python
import random
inventory={}
total_money=0
fish_prices={"rohu":12,"catla":18,"sinchu":20}
print("\n 🎣🐟fishing game")
print("1 catch the fish")
print("2 view inventory")
print("3 sell fish")
print("4 quit")
while True:
  choice=int(input("choose an option:"))
  if choice==1:
   fish=random.choice(list(fish_prices.keys()))
   inventory[fish]=inventory.get(fish,0)+1
   print(f"you caught a {fish}")
  elif choice==2:
   if inventory=={}:
     print("no fish")
   else:
      print("your inventory:")
   for fish,count in inventory.items():
      print(f"{fish}:{count}")
  elif choice==3:
    if inventory=={}:
      print("no fish sell")
    else:earnings =0
    for fish,count in inventory.items():
      earnings+=fish_prices[fish]*count
    total_money+=earnings
    inventory.clear()
    print(f"you sold {count} {fish} for ${earnings}")
  elif choice==4:
    print(f"final moneynearned:${total_money}")
    print("thank yoy or playing🙏")
    break
```
