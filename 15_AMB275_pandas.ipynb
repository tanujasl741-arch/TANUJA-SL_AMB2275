```python
import pandas as pd
production=pd.Series([30,60,45,25],
                     index=["wheat","rice","maize","barley"]
)
print(production)
```

```python
#by crop name label
print(production['rice'])
print(production.loc["rice"])
```

```python
#by position
print(production[1])
print(production.iloc[1])
```

```python
print(production.sum())
print(production.mean())
print(production.max())
print(production.min())
```

```python
#increase all crop production by 10%(good harvest year)
increased_production=production*1.1
print(increased_production)
```

```python
#crops with production greater than 40tons
high_production=production[production>40]
print(high_production)
print()

print(production>40)
```

```python
import pandas as pd
data={
    "farm":["farm1","farm2","farm3","farm4"],
    "crop": ["wheat","rice","maize","barley"],
    "area_ha":[10,12,14,15],
    "production_tons":[30,60,24,36]
}
df=pd.DataFrame(data)
print(df)
```

```python
#add a new column
#yld=production by area
df.yld_per_ha=df.production_tons/df.area_ha
print(df)
```

```python
df["yld_per_ha"]=df["production_tons"]/df["area_ha"]
print(df)
```

```python
print(df.iloc[[1,3]])
```

```python
print(df.iloc[[1,3]],[0,1,2])
```

```python
print(df.loc[[1,3],["farm","crop","area_ha"]])
```

```python
#filter data
#farms growing wheat
print(df["crop"]=="rice")
print()
wheat_farms=df[df["crop"]=="wheat"]
print(wheat_farms)
```

```python
#farms with yield greater than 3
high_yld=df[df["yld_per_ha"]>3]
print(high_yld)
```

```python
#sort farms by highest production
sorted_df=df.sort_values(by="production_tons",ascending=False)
print(sorted_df)
```

```python
#group operation avg yld per crop
avg_yld=df.groupby("crop")["yld_per_ha"].mean()
print(avg_yld)
```

```python
avg_yld=df.groupby("crop")["yld_per_ha"].sum()
print(avg_yld)
```
