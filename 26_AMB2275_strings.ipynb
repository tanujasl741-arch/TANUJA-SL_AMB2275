```python
a="today is 19-jan-2026 monday"
B=C=V=D=O=0
for i in a:
  if i==" ":
    B=B+1

  elif i in "aeiou":
   V=V+1
  elif i.isalpha():
    C=C+1
  elif i in "bcdfghjklmnpqrstvwxyz":
    C=C+1

  elif i.isdigit():
    D=D+1

  else :
    O=O+1
print(B)
print(V)
print(C)
print(D)
print(O)
```
