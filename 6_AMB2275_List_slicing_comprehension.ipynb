Basic List Slicing

Syntax: list[start : end : step]

End index is excluded <br>
Slicing does not modify the original list

```python
numbers = [10, 20, 30, 40, 50, 60]
print(numbers[0:3])   # First three elements
```

```python
print(numbers[2:5])   # From index 2 to 4
print(numbers[:4])    # From beginning to index 3
print(numbers[3:])    # From index 3 to end
print(numbers[7])   # error: list index out of range
```

```python
numbers[-3:-1]     #-3  -2  -1
```

```python
numbers[-3:-5]
```

```python
numbers[5:3]   # 5  6
```

```python
numbers = [10, 20, 30, 40, 50, 60]
         #  0  1   2   3   4   5
print(numbers[1:6])
print(numbers[1:6:2])
print(numbers[1:6:1])
print(numbers[1:6:])
print(numbers[1:6:3])   # 20 50
```

```python
numbers = [10, 20, 30, 40, 50, 60]
          # -6  -5  -4  -3  -2  -1
print(numbers[-5:-1])
print(numbers[-5:-1:2])   # -5 -3 -1  -> -1 not included
print(numbers[-5:-1:1])
print(numbers[-5:-1:])
print(numbers[-5:-1:3])   # 20 50

print(numbers[::2])   # Every second element
print(numbers[1::2]) # Elements at odd index
```

```python
numbers = [10, 20, 30, 40, 50, 60]
          # -6  -5  -4  -3  -2  -1
numbers[-3:-5:-1]
```

```python
crops = ["Rice", "Wheat", "Maize", "Millet", "Pulses", "Cotton"]
        # -6      -5       -4      -3       -2       -1

print(crops[:3])      # First three crops
print(crops[1:5])     # Middle crops, excluding end crops
print(crops[-5:-1])     # Middle crops, excluding end crops
print(crops[-2:])     # Last two crops
```

```python
#Alternate Crops
crops = ["Rice", "Wheat", "Maize", "Millet", "Pulses", "Cotton"]
print(crops[::2])    # Alternate crops
```

```python
print(crops[::-1])   # Reverse a List Using Slicing
```

```python
crops = ["Rice", "Wheat", "Maize", "Millet", "Pulses", "Cotton"]
print(crops[1:-1])   # Remove first and last crop
print(crops[1:5])
print(crops[-5:-1])
```

```python
# list comprehension
numbers = [10, 20, 30, 40, 50, 60]
for i in numbers:
    print(i+2)
print(numbers)
```

```python
# create a new list with each element increased by 2
numbers = [10, 20, 30, 40, 50, 60]
n=[]
for i in numbers:
    n.append(i+2)
print("new list",n)
print("original list",numbers)
```

```python
#list comprehension - create a new list with each element increased by 2
numbers = [10, 20, 30, 40, 50, 60]
        # n=[ expression for item in iterable ]
n=[i+2 for i in numbers]
print("new list",n)
print("original list",numbers)
```

```python
#create a new list with numbers divisible by 4
numbers = [10, 20, 30, 40, 50, 60]
n=[]
for i in numbers:
    if i%4==0:
        n.append(i)
print("numbers divided by 4",n)
print("original list",numbers)
```

```python
#create a new list with numbers divisible by 4 using list comprehension
numbers = [10, 20, 30, 40, 50, 60]
         # n=[expression for item in iterable if condition]
n=[i for i in numbers if i%4==0]
print("numbers divided by 4",n)
print("original list",numbers)
```

```python
# print even or odd using list comprehension and if else
numbers = [10, 21, 30, 41, 50, 61]
n=["even" if i%2==0 else "odd" for i in numbers]

#n=[expression1 if condition else expression2 for item in iterable]
print(n)
```
