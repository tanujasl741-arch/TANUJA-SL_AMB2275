```python
import numpy as np
crop_yield=np.array([
    [12,13,14,15,16],
    [11.2,12.2,11.3,12,14],
    [13,16,2,3,4],
    [8.9,3,5,5.6,8.4]
 ])
print('crop yield array(tons):')
print(crop_yield)
print(f'\nshape:{crop_yield}')
print(f'data type:{crop_yield.shape}')
print(f'total elements:{crop_yield.size}')

print(f'number of dimensions:{crop_yield.ndim}')
```

```python
import numpy as np
total_yield=np.sum(crop_yield)
avg_yld=np.mean(crop_yield,axis=1)
avg_yld_perseason=np.mean(crop_yield,axis=0)
maxyld=np.max(crop_yield)
minyld=np.min(crop_yield)



print(f'total yield:{total_yield}')
print(f'average yield:{avg_yld}')
print(f'average yield per season:{avg_yld_perseason}')
print(f'maximum yield:{maxyld}')
print(f'minimum yield:{minyld}')
```

```python
#find best performing farm
best_farm=np.argmax(avg_yld)
worst_farm=np.argmin(avg_yld)
#arg max()does not return the max value itself it also gives value where it i slocated
print('\nbest performing farm:',best_farm+1)
print(f'their avg yield:{avg_yld[best_farm]}')
print('\nworst performing farm:',worst_farm+1)
```

```python
#boolean indexing :find exceptional yields
exceptional_yields=crop_yield[crop_yield>13]
print(exceptional_yields)
print(f'exceptional yields:{exceptional_yields}')
```

```python
#conditional operators:classify yields
categories=np.where(crop_yield>13,'high','low')
print(f'farm categories:\n{categories}')
```
