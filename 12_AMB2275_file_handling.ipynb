```python
def expr(a,b):
  try:
    c=a/b
    print(c)
  except ZeroDivisionError:
   print("division by zero")
  except NameError:
    print("variable not defined")
  except:
    print("some issue")
  finally:
    print("End of the program")
a=2
b=33
expr(a,b)
```
