Example 8: Fertilizer Recommendation Using Two Lists

```python
crops = ["Rice", "Wheat", "Maize", "Pulses"]
fertilizer = ["Urea", "DAP", "NPK", "Compost"]

print("Recommended fertilizer:")
for i in range(len(crops)):
    print(crops[i], "→", fertilizer[i])

print("Recommended fertilizer:")
for crop, fert in zip(crops, fertilizer):
    print(crop, "→", fert)
```

```python
# 18. Top three most expensive equipment
# Combine equipment and cost

equipment = ["Tractor", "Plough", "Harvester", "Seeder", "Sprayer", "Thresher"]
cost = [850000, 25000, 1200000, 60000, 18000, 95000]

# Step 1: Copy and sort cost list in descending order
sorted_cost = cost.copy()
sorted_cost.sort(reverse=True)

# Step 2: Take top three costs
top_three_costs = sorted_cost[:3]

# Step 3: Find corresponding equipment
top_three = []

for c in top_three_costs:
    index = cost.index(c)
    top_three.append((equipment[index], c))

print("Top three most expensive equipment:", top_three)
```

```python
# 18. Top three most expensive equipment

equipment = ["Tractor", "Plough", "Harvester", "Seeder", "Sprayer", "Thresher"]
cost = [850000, 25000.15, 1200000, 60000, 18000, 95000]

# Combine cost and equipment (cost first)
items = []

for i in range(len(equipment)):
    items.append([cost[i], equipment[i]])

# Sort in descending order by cost
items.sort(reverse=True)

# Take top three
top_three = items[:3]

print("Top three most expensive equipment:")
for item in top_three:
    print(item[1], "-", item[0])
```

```python
# 19. Reduce cost by 15% for subsidy-eligible equipment
# (Assuming subsidy for equipment costing below ₹1,00,000)
subsidy_cost = []
for c in cost:
    if c < 100000:
        subsidy_cost.append(int(c * 0.85))
    else:
        subsidy_cost.append(c)
print("Costs after subsidy:", subsidy_cost)

subsidy_cost = [
    int(c * 0.85) if c < 100000 else c
    for c in cost
]
print("Costs after subsidy:", subsidy_cost)
```
