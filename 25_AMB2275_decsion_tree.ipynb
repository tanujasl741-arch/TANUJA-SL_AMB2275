```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import pandas as pd
#read the data set
data=pd.read_csv('iris.csv')
print(data)
```

```python
data["Species"].unique()#gives dataframe["column name"] gives unique values in the column
```

```python
data["Species"].value_counts()#gives count of each unique value in the column name
```

```python
#separate dependent and independent variable (input/output)
x=data.iloc[:,:-1] # All columns except the last one (features)
y=data.iloc[:,-1] # The last column (target variable)

from sklearn.model_selection import train_test_split
x_train,x_test,ytrain,y_test=train_test_split(x,y,test_size=1/3,random_state=0)
```

```python
#create descion tree classifer object
from sklearn.tree import DecisionTreeClassifier
classifier=DecisionTreeClassifier(criterion="entropy",max_depth=None,min_samples_split=5,min_samples_leaf=2)
```

```python
#train model('fit' method)
model=classifier.fit(x_train,ytrain)
print(model)
```

```python
# predict the test model
y_pred=model.predict(x_test)
print(y_pred)
```

```python
from sklearn.metrics import confusion_matrix
c=confusion_matrix(y_test,y_pred)
c
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_pred))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python


#visualisation
from sklearn.tree import plot_tree
fn=['sepal length(cm)','sepal width(cm)','petal length(cm)','petal width(cm)' ]
cn=['septosa','versicolor','virginica']
fig,axes=plt.subplots(nrows=1,ncols=1,figsize=(4,4),dpi=300)
plot_tree(model,
               feature_names=fn,
               class_names=cn,
               filled=True);
```
